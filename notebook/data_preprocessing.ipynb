```python
sales["sub_category"] = sales["sub_category"].fillna("Unknown")
```

```python
sales["profit_margin_pct"] = (
    (sales["selling_price"] - sales["unit_cost"] - sales["fba_fees"])
    / sales["selling_price"]
) * 100
```

```python
sales.isnull().sum()
```

```python
ads.isnull().sum()
```

```python
sales["date"] = pd.to_datetime(sales["date"])
ads["date"] = pd.to_datetime(ads["date"])
```

```python
ads["acos"] = ads["ad_spend"] / ads["attributed_sales"]
ads["acos"] = ads["acos"].replace([np.inf, -np.inf], 0)
ads["acos"].fillna(0, inplace=True)
```

```python
ads["roas"] = ads["attributed_sales"] / ads["ad_spend"]
ads["roas"] = ads["roas"].replace([np.inf, -np.inf], 0)
ads["roas"].fillna(0, inplace=True)
```

```python
sales["profit_per_unit"] = sales["selling_price"] - sales["unit_cost"] - sales["fba_fees"]
```

```python
sales = sales.sort_values("date")
ads = ads.sort_values("date")
```

```python
sales.to_csv("D:/amazon_agent_ai/data/sales_cleaned.csv", index=False)
ads.to_csv("D:/amazon_agent_ai/data/ads_cleaned.csv", index=False)
```
